# Apply all models

```python
ran_state = 42 
metrics_df = None

# give dataset name
dataset_name = 'ex8data1'

# set the classifiers
classifiers = {
    'HBOS': HBOS(),
    'I-Forest': IForest(behaviour='new', random_state=ran_state),
    'KNN': KNN(),
    'LOF': LOF(),
    'MCD': MCD(random_state=ran_state),
    'OCSVM': OCSVM(),
    'Perception': Perception(),
    #'DBSCAN' : DBSCAN(), # ignored due to its poor performance on every other dataset
}

# run the classifiers over the data and produce results
for i, (clf_name, clf) in enumerate(classifiers.items()):

    if clf_name == 'DBSCAN':
        t0 = time.time()
        clustering = clf.fit(X_val)
        labels = np.where(clustering.labels_==-1, 1, 0)
        t1 = time.time()
        train_time = t1-t0
        predict_time = 0
        total_time = t1-t0
                
    else:
        # fit to training data
        t0 = time.time()
        clf.fit(X_val)
        t1 = time.time()
        train_time = t1-t0

        # predict on test data
        t0 = time.time()
        labels = clf.predict(X_val)
        t1 = time.time()
        predict_time = t1-t0
        
        total_time = train_time + predict_time
            
    cls_report_dic = classification_report(y_val, labels, 
                                       target_names=['normal', 'abnormal'],output_dict=True)

    if clf_name == 'DBSCAN':
        # show binary label version of auc score
        auc_score = roc_auc_score(y_val, labels)
    elif clf_name =='Perception':
        auc_score = roc_auc_score(y_val, clf.scores_)
    else:
        auc_score = roc_auc_score(y_val, clf.decision_scores_)

    # collect all the metrics
    metrics_temp = pd.DataFrame({
                    'Dataset': [dataset_name],
                    'Classifier': [clf_name],
                    'Precision': [cls_report_dic['abnormal']['precision']], 
                    'Recall': [cls_report_dic['abnormal']['recall']],
                    'F1-score': [cls_report_dic['abnormal']['f1-score']],
                    'Support': [cls_report_dic['abnormal']['support']],
                    'AUC-score': [auc_score],
                    'Train time': [train_time],
                    'Predict time': [predict_time],
                    'Total time': [total_time],
                   })

    metrics_df = pd.concat([metrics_df, metrics_temp]).reset_index(drop=True)

list_round_2 = ['Precision', 'Recall', 'F1-score', 'AUC-score']
list_round_5 = ['Train time', 'Predict time', 'Total time']
metrics_df[list_round_2] = metrics_df[list_round_2].round(2)
metrics_df[list_round_5] = metrics_df[list_round_5].round(5)
metrics_df
```

```python
columns_to_drop = ['Dataset', 'Train time', 'Predict time', 'Support']
metrics_df_dropped = metrics_df.drop(columns_to_drop, axis=1)

df = metrics_df_dropped.copy()
df
```

```python
cols = ['Precision', 'Recall', 'F1-score', 'AUC-score']
formatdict = {}
for col in cols: formatdict[col] = "{:.2f}"
formatdict.pop('Classifier', None)
formatdict['Total time'] = "{:.5f}"
    
metrics_df_styled = df.style.hide_index().apply(highlight_max, 
                                                 subset=['Precision', 'Recall', 'F1-score', 
                                                         'AUC-score']).\
apply(highlight_min, subset=['Total time']).format(formatdict)
metrics_df_styled
```
